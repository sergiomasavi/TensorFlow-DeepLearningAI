# El HelloWorld del Deep Learning con Redes Neuronales

En este ejemplo veremos como ajustar números a una línea para demostrar cómo funciona el Deep Learning. Para ello, utilizaremos un conjunto de números de entrada (X) para tratar de encontrar su valor de salida (y), es decir, calcular la fórmula que mapea X en Y. Esta función es:

```
def hw_function(x):
    y = (2 * x) - 1
    return y
```

## Librerías

En este ejemplo se importa:
- [TensorFlow](https://www.tensorflow.org/) y se llama `tf` por convención y facilidad de uso.
- [numpy](https://numpy.org) que ayuda a representar datos como arrays fácilmente y a optimizar las operaciones numéricas.
- [keras](https://keras.io/) es el framework que se utiliza para construir redes neuronal en TensorFlow como secuencias es 

````python
import tensorflow as tf
from tensorflow import keras as ks

print('TensorFlow:', tf.__version__)
print('Keras:',ks.__version__)
````

## Definición y compilación de la red neuronal

Creamos la red neuronal más sencilla posible de 1 capa con 1 neurona, y la forma de entrada a la misma es sólo 1 valor. Para construir este modelo se utiliza:
- [Sequential](https://keras.io/api/models/sequential/) de Keras que permite definir la red como una secuencia de - - - [layers](https://keras.io/api/layers/). 
- [Dense](https://keras.io/api/layers/core_layers/dense/) para construir esta red simple.

````python
# Definir neurona
model = ks.Sequential([ks.layers.Dense(units=1, input_shape=[1])])
````

A continuación, se compila la red neuronal especificando dos funciones:
- [loss](https://keras.io/api/losses/). La función "pérdida" mide lo bien o lo mal que se calcularon las predicciones contra los valores reales conocidos.
- [optimizer](https://keras.io/api/optimizers/). Función que basándose en cómo fue la función de pérdida, tratará de minimizar la pérdida.

El bucle de calcular la función de pérdida y ejectuar el optimizador para optimizar su valor se repetirá para el número de __epochs__ establecido en el entrenamiento. En este ejemplo, se utilizará:
- [error cuadrático medio](https://keras.io/api/losses/regression_losses/#meansquarederror-function) para la pérdida.
- [descenso de gradiente estocástico](https://keras.io/api/optimizers/sgd/) para el optimizador.

````python
model.compile(optimizer='sgd', loss='mean_squared_error')
model.summary()
````

## Datos de entrenamiento

A continuación, se definen los datos que se utilizarán en el entrenamiento. En este caso, la relación entre las variables (X,y) es $y=2x-1$..

````python
# Datos de entrenamiento
X = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
print('X=',X)
print('y=',y)
````

## Entrenamiento
El proceso de entrenamiento de la red neuronal, en el que "aprende" la relación entre las x y las y, se realiza con el método __[model.fit()](https://keras.io/api/models/model_training_apis/#fit-method).__ Aquí es donde se ejectua el bucle mencionado anteriormente que consiste en:
- Hacer una suposición.
- Medir lo buena o mala que es.
- Usar el opimizador para hacer otra conjetura, etc. 

Este proceso se repetirá durante el número de `epocas` que se especifiquen.

````python
# Entrenamiento
print('Training...')
history = model.fit(X, y, epochs=500, verbose=True)
````

Una vez el modelo ha sido entrenado, ha ajustado sus pesos internos y aprendido cuál es la relación entre __X__ e __y__. A continuación, se utiliza el método __[model.predict](https://keras.io/api/models/model_training_apis/#predict-method)__ para que calcule el valor __y__ para una __X__ previamente desconocida.

````python
print('Predict X=10. y=',model.predict([10.0]))
````

Se observa que el valor es muy cercano a 19 pero no exactamente 19. Esto se debe fundamentalmente a dos razones principales:
- El número de datos de entrenamientos es muy bajo. Solamente hay seis puntos. Esos seis puntos son lineales pero no hay garantía de que por cada X, la relación será Y igual 2X menos 1. Hay una gran probabilidad de que Y sea igual a 19 para X igual a 10, pero la red neuronal no es positiva. 

- En consecuencia a la razón anterior, la red neuronal averiguará un valor realista para Y, es decir, como averiguan la respuesta para todo, trabajan con probabilidades.
